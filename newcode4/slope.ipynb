```python
data={}
data['E288 200'] = np.loadtxt("data/reformat-E288-200.dat", skiprows=2)
data['E288 300'] = np.loadtxt("data/reformat-E288-300.dat", skiprows=2)
data['E288 400'] = np.loadtxt("data/reformat-E288-400.dat", skiprows=2)
data['E605 800'] = np.loadtxt("data/reformat-E605-800.dat", skiprows=2)
data['E772 800'] = np.loadtxt("data/reformat-E772-800.dat", skiprows=2)
```

# Slope

## Fit function

```python
def func(x, a, b):
    return np.exp(-x * a + b)
```

```python
dset = sdata['E288 200']
qts = np.unique(dset[:,1]) 

nrows, ncols = len(qts) // 4 + 1, 4
fig = py.figure(figsize=(ncols*7, nrows*7))
for i in range(len(qts)):
    ax = fig.add_subplot(nrows, ncols, i+1)
    qt = qts[i]
    sub = np.logical_and(dset[:,1] == qt, dset[:,0] != 9.5, dset[:,0] != 10.5)
    X = dset[sub, 0] / dset[sub, 4]
    Y = dset[sub, 2]
    Yerr = dset[sub, 3]
    ax.errorbar(X, Y, yerr=Yerr, marker='o', markersize=15, linewidth=2, linestyle='none', color='black')
    ax.tick_params(axis='both', which='both', direction='in', labelsize=25)
    ax.set_yscale('log')
    ax.set_xlim(0.15, 0.7)
    #ax.set_ylim(1e-15, 1e-9)
    ax.set_xlabel(r'$Q/\sqrt{s}$', size=25)
    ax.set_ylabel(r'$\langle \frac{d^3\sigma}{dydq_T^2dQ^2}\rangle~(\rm fm^2/GeV^4)$', size=25)
    ax.text(0.6, 0.9, r'$q_T=%.1f~\rm GeV$' %(qt), transform = ax.transAxes, size = 25)
    if len(X) > 2:
        par, cov = curve_fit(func, X, Y, sigma=Yerr)
        Xc = np.linspace(0.15, 0.7, 20)
        ax.plot(Xc, func(Xc, par[0], par[1]), linestyle='-', linewidth=2, color='blue')
        ax.text(0.6, 0.8, r'$a=%.2f$' %(par[0]), transform = ax.transAxes, size = 25)
fig.text(0.5, 1.01, r'E288 200 GeV $\sqrt{s}=19.4~\rm GeV$', horizontalalignment='center', fontsize=40)        
fig.tight_layout()
fig.savefig("plots/slope-E288-200.pdf")
```
